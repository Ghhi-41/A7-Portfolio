# Working With Data 2

```python
import pandas as pd, os
from google.colab import drive
drive.mount('drive')
```

```python
fpath = 'drive/MyDrive/ANLY 4100/DataSets/retail_data/'
os.listdir(fpath)
```

```python
storesA = pd.read_csv(fpath+'stores_A_data-set.csv', index_col = 0)
storesB = pd.read_csv(fpath+'stores_B_data-set.csv', index_col = 0)
storesC = pd.read_csv(fpath+'stores_C_data-set.csv', index_col = 0)
storesA.shape[0],storesB.shape[0],storesC.shape[0] # number of rows in each
```

```python
storesA.head()
```

```python
storesB.head()
```

```python
storesC.head()
```

```python
stores = pd.concat([storesA, storesB, storesC])
stores.reset_index(drop = True, inplace = True)
stores.Type.unique()
```

```python
stores.shape
```

```python
stores.head()
```

```python
stores.to_csv(fpath +'stores_data-set')
```

```python
stores = pd.read_csv(fpath+'stores_data-set.csv', index_col = 0) # we can read the combined dataset
stores.shape
```

```python
sales = pd.read_csv(fpath+'sales_data_set.csv') # read in the sales dataset
sales.shape
```

```python
sales.head()
```

```python
sales.Date.dtype
```

```python
sales.Date = pd.to_datetime(sales.Date)
```

```python
sales.Date.dtype
```

```python
sales.head()
```

```python
df = sales.merge(stores, how = 'left', on = 'Store')
df.shape
```

```python
df.head()
```
